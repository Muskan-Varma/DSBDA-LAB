Perform the following operation using titanic data set.
1. check how the price of the ticket (column name: &#39;fare&#39;) for each
passenger is distributed by plotting a histogram.
2. plot a box plot for distribution of age with respect to each gender along
with the information about whether they survived or not. (Column
names : &#39;sex&#39; and &#39;age&#39;)
3. Write observations on the inference from the above statistics.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_df = pd.read_csv('Titanic.csv')
```

```python
# 1. Check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram.
plt.figure(figsize=(8, 4))
sns.histplot(data=titanic_df, x='Fare', bins=20, kde=True)
plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Plot a box plot for the distribution of age with respect to each gender along with the information about whether they survived or not. (Column names: 'sex' and 'age')
plt.figure(figsize=(8, 4))
sns.boxplot(data=titanic_df, x='Sex', y='Age', hue='Survived')
plt.title("Distribution of Age with Respect to Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python
# 3. Observations:
# - From the histogram of ticket fares, we can observe the distribution of ticket prices paid by passengers.
# - From the box plot, we can observe the distribution of age with respect to gender and survival status. 
#   - It appears that on average, females tend to be younger than males.
#   - Additionally, the box plot shows the age distribution for both survived and not survived passengers. We can observe that the median age of survived passengers tends to be slightly lower than that of non-survived passengers, especially for females.
```
