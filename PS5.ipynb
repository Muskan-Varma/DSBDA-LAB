Perform the following operations on iris dataset
1. List down the features and their types
2. Create a box plot and histogram for each feature in the dataset.
3. Compare distributions and identify outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
# iris = load_iris()

iris=pd.read_csv("Iris.csv")
iris
```

```python
# Create a DataFrame from the dataset
iris_df = pd.DataFrame(iris)
```

```python
# 1. List down the features and their types
feature_types = iris_df.dtypes
print("Features and their types:")
print(feature_types)
```

```python
# 2. Create a box plot and histogram for each feature in the dataset
# Box plot
plt.figure(figsize=(15, 10))
for i, feature in enumerate(iris.columns[1:]):
    plt.subplot(3, 2, i + 1)
    sns.boxplot(x='Species', y=feature, data=iris_df)
    plt.title(f"Boxplot of {feature}")
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(15, 10))
for i, feature in enumerate(iris.columns[1:]):
    plt.subplot(3, 2, i + 1)
    sns.histplot(data=iris_df, x=feature, bins=20, kde=True)
    plt.title(f"Histogram of {feature}")
plt.tight_layout()
plt.show()
```

```python
# 3. Compare distributions and identify outliers
# From the box plots and histograms, we can compare the distributions of each feature among different species of iris.
# Outliers can be identified by observing points that fall significantly outside the whiskers of the box plot.
```
